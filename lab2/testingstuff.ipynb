```python
def lms(arr1, arr2):
    len1 = len(arr1)
    len2 = len(arr2)
    max_len = 0
    end = -1
    for i in range(len1):
        for j in range(len2):
            k = 0
            while (i + k < len1) and (j + k < len2) and (arr1[i + k] == arr2[j + k]):
                k += 1
            if k > max_len:
                max_len = k
                end = i + k - 1
    return arr1[end - max_len + 1 : end + 1]

def overlap(s1, s2):
    a1 = [word.lower().strip(".,;:!?") for word in s1.split()]
    a2 = [word.lower().strip(".,;:!?") for word in s2.split()]
    k = []
    while lms(a1, a2):
        k.append(len(lms(a1, a2)))
        for i in lms(a1, a2):
            a1.remove(i)
    overlap = 0
    for i in k:
        overlap += i**2
    return overlap
```

```python
def get_hyponyms(synset):
    hyponyms = set()
    for hyponym in synset.hyponyms():
        hyponyms |= set(get_hyponyms(hyponym))
    return hyponyms | set(synset.hyponyms())
```

```python
def count(syn):
    c = 0
    a = get_hyponyms((syn))
    a.add(syn)
    for i in a:
        for j in i.lemmas():
            c += j.count() + 1
    return c
```
